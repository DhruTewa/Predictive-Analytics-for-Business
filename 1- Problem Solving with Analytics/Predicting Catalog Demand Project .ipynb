> 2.	How did you come up with your recommendation? 

- Step 1- Calculate the predicted sales from each customer.
- Step 2- Multiply the predicted sales by score_yes probability value.
- Step 3- As the average gross margin) on all products sold through the catalog is 50%, multiply the value we got in Step 2 by 50%.
- Step 4- Subtract the value of catalog from the value we get in step 3.
- Step 5- The value we get from the Step 4 is the total profit gained from each product.
- Step 6- By adding all the profit gained we get the value higher that \\$10,000 hence recommended to send out the catalog.
